```python
from ultralytics import YOLO

# Load a COCO-pretrained YOLO12n model
model = YOLO("yolo12s.pt")

# # Train the model on the COCO8 example dataset for 100 epochs
# results = model.train(data="/kaggle/input/pedestrian-detection-fogg-and-rain/dataset/dataset/data.yaml", epochs=50, imgsz=640, batch=16)
results = model.train(data="/kaggle/input/pedestrian-detection-fogg-and-rain/dataset/dataset/data.yaml",
epochs=50,              # total training epochs
    batch=16,               # auto batch size (fits max on GPU)
    lr0=0.01,               # initial learning rate
    device='0,1',           # use both available GPUs
    imgsz=640,              # image size
    amp=True,               # use mixed precision for speed/memory
    save=True,              # save best model automatically
    project='runs/train',   # folder where training logs & weights are saved
    name='yolov12_pedestrian'  # name of your training run
                     )
```
