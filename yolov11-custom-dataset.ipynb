```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11s.pt")  # load a pretrained model (recommended for training)

# Train the model
results = model.train(data='/kaggle/input/pedestrian-detection-fogg-and-rain/dataset/dataset/data.yaml',
    epochs=50,                 # fits within GPU 12hr window
    batch=16,                  # compatible with T4 x2
    imgsz=640,                 # input image size
    device='0,1',              # use both T4 GPUs
    amp=True,                  # mixed precision = faster training
    lr0=0.01,                  # initial learning rate
    project='runs/train',
    name='yolov11_pedestrian',
    save=True)
```

```python
import os
import random
from ultralytics import YOLO
from PIL import Image
import shutil

# Path to your test folder
test_folder = '/kaggle/input/pedestrian-detection-fogg-and-rain/dataset/dataset/test/images'
images = [f for f in os.listdir(test_folder) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

# âœ… Sample 500 unique images randomly
sample_images = random.sample(images, 500)

# Load YOLO model
model = YOLO('runs/train/yolov11_pedestrian/weights/best.pt')

# Optional: Create output folder
output_dir = 'runs/detect/sample_predict_random_500'
os.makedirs(output_dir, exist_ok=True)

# Loop through random images and run prediction
for i, img_name in enumerate(sample_images):
    img_path = os.path.join(test_folder, img_name)

    results = model.predict(
        source=img_path,
        conf=0.55,
        save=True,           # saves to runs/detect/
        save_txt=True,
        imgsz=640
    )

    print(f"[{i+1}/500] Predicted {img_name}")
```
